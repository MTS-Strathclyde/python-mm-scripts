```python
def get_uc_from_results(rism_calc):
    results = rism_calc.Results.splitlines()
    return results[-1].split()[1]

def average(l):
    summ = 0
    for i in l:
        summ += float(i)
    try:
        return summ/len(l)
    except ZeroDivisionError:
        return None
    
def get_298_exp_from_mol(mol):
    experiments = mol.Experiments
    def filtr_temperature_298(exp): return True if 298 <= exp.Temperature < 299 else False
    experiments = filter(filtr_temperature_298, experiments)
    if len(experiments) == 0:
        return None
    experiment = experiments[0]
    return experiment.SolvEnergy
    
```

```python
results = []

for conf in ses.query(Conformation).join(Molecule).filter(Molecule.Experiments != None):
    res_line = []
    res_line.append(conf.Molecule_INCHI)
    res_line.append(conf.molecule.IUPACName)
    res_line.append(get_298_exp_from_mol(conf.molecule))
    calcs = [(), (), ()]
    for calc in conf.Calculations:
        if calc.CalculationType == 'CMCalculation':
            if calc.SolvMethod == 'SMD':
                calcs[1] += (calc.SolvE,)
            elif calc.SolvMethod == 'SMVLE':
                calcs[2] += (calc.SolvE,)
        elif calc.CalculationType == 'RISMCalculation':
            calcs[0] += (get_uc_from_results(calc),)
    calcs = map(average, calcs)
    res_line.extend(calcs)    
    if None not in res_line:
        results.append(res_line)
            
```
