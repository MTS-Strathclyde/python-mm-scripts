```python
from solvation_database import Molecule, Conformation, Experiment
from solvation_database import Calculation, RISMCalculation, CMCalculation
from db_interface import DBInterface, create_session
```

```python
import csv
import pandas
```

```python
ses = create_session()
```

```python
data_rows = []
```

```python
molecules = ses.query(Molecule).all()
print len(molecules)
```

```python
for molecule in molecules:
    rows = [[molecule.SMILES , molecule.IUPACName]]
    experiments = molecule.Experiments
    exp_rows = []
    def filtr_temperature_298(exp): return True if 298 <= exp.Temperature < 299 else False
    experiments = filter(filtr_temperature_298, experiments)
    if len(experiments) == 0:
        continue
    for experiment in experiments:
        exp_rows.append(['exp', experiment.Temperature, experiment.SolvEnergy])
    rows.extend(exp_rows)
    confs = molecule.Conformations
    if not confs:
        continue
    for conf in confs:
        for calc in conf.Calculations:
            rows.append([calc.CalculationType, calc.Temperature, calc.SolvE])
    data_rows.extend(rows)
    
```

```python
with open('del_me', 'wb') as f:
    wr = csv.writer(f)
    wr.writerows(data_rows)
```

```python
df = pandas.read_csv('del_me',header=None, names=['SMILES/Type', 'IUPAC/Temp', 'None/Energy'])
```
